# Sales Data Analysis Project
This project analyzes sales performance using Python (Pandas, Matplotlib, Seaborn).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    "Date": pd.date_range(start="2024-01-01", periods=10, freq='D'),
    "Product": ["Phone", "Laptop", "Tablet", "Phone", "Laptop", "Tablet", "Phone", "Laptop", "Tablet", "Phone"],
    "Region": ["East", "West", "East", "North", "West", "East", "North", "East", "West", "North"],
    "Sales": [500, 800, 300, 450, 900, 350, 600, 700, 400, 650]
}

df = pd.DataFrame(data)
df
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
df.fillna(0, inplace=True)
```

```python
df["Month"] = df["Date"].dt.month
df
```

```python
df["Sales"].sum()
```

```python
df.groupby("Product")["Sales"].sum()
```

```python
df.groupby("Month")["Sales"].sum()
```

```python
df.groupby("Product")["Sales"].sum().plot(kind="bar")
plt.title("Sales by Product")
plt.show()
```

```python
df.groupby("Region")["Sales"].sum().plot(kind="bar")
plt.title("Sales by Region")
plt.show()
```

```python
df.groupby("Month")["Sales"].sum().plot(kind="line")
plt.title("Monthly Sales Trend")
plt.show()
```
