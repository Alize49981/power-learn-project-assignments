```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = {
    "Name": ["Ali", "John", "Asha", "Mary"],
    "Math": [85, 78, 92, 88]
}

df = pd.DataFrame(data)

sns.barplot(x="Name", y="Math", data=df)

plt.title("Math Scores")
plt.show()
```

```python
sns.histplot(df["Math"], bins=5)
plt.show()
```

```python
import pandas as pd

data = {
    "Name": ["Ali", "John", "Asha", "Mary", "Peter"],
    "Gender": ["M", "M", "F", "F", "M"],
    "Math": [85, 78, 92, 88, 74],
    "Science": [90, 85, 88, 76, 80],
    "English": [78, 80, 91, 85, 79]
}

df = pd.DataFrame(data)
df
```

```python
df.fillna(0, inplace=True)
```

```python
df["Total"] = df["Math"] + df["Science"] + df["English"]
df["Average"] = df["Total"] / 3
```

```python
df_sorted = df.sort_values(by="Total", ascending=False)
df_sorted
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x="Name", y="Total", data=df_sorted)
plt.title("Student Ranking by Total Score")
plt.show()
```

```python
sns.boxplot(x="Gender", y="Total", data=df)
plt.title("Score Distribution by Gender")
plt.show()
```

```python
sns.heatmap(df[["Math","Science","English","Total"]].corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```
