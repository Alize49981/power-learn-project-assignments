```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris

# Load dataset
iris = load_iris()

# Convert to DataFrame
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Add target column (species)
df['species'] = iris.target

# Map numeric species to names
df['species'] = df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
```

```python
print(df.head())
```

```python
print(df.info())        # Data types
print(df.isnull().sum())  # Missing values
```

```python
# Fill missing values (if any)
df.fillna(df.mean(numeric_only=True), inplace=True)

# OR drop missing values
# df.dropna(inplace=True)
```

#Task 2: Basic Data Analysis

```python
print(df.describe())
```

```python
grouped = df.groupby('species').mean()
print(grouped)
```

#Task 3: Data Visualization

```python
plt.figure()
df['sepal length (cm)'].plot(kind='line')
plt.title('Sepal Length Trend')
plt.xlabel('Index')
plt.ylabel('Sepal Length (cm)')
plt.legend(['Sepal Length'])
plt.show()
```

```python
grouped['petal length (cm)'].plot(kind='bar')
plt.title('Average Petal Length per Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.show()
```

```python
plt.hist(df['sepal width (cm)'], bins=10)
plt.title('Distribution of Sepal Width')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.scatter(df['sepal length (cm)'], df['petal length (cm)'])
plt.title('Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```

```python
sns.scatterplot(data=df, x='sepal length (cm)', y='petal length (cm)', hue='species')
plt.title('Seaborn Scatter Plot')
plt.show()
```
